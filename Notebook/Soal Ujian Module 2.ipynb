### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
man = dataTitanic.groupby('who')['age'].mean()['man']
child = dataTitanic.groupby('who')['age'].mean()['child']
woman = dataTitanic.groupby('who')['age'].mean()['woman']
lenNanMan = len(dataTitanic[(dataTitanic['who'] == 'man') & (dataTitanic['age'].isna())])
lenNanChild = len(dataTitanic[(dataTitanic['who'] == 'child') & (dataTitanic['age'].isna())])
lenNanWoman = len(dataTitanic[(dataTitanic['who'] == 'woman') & (dataTitanic['age'].isna())])
print(f'Banyak NaN man & mean: {lenNanMan} & {man}')
print(f'Banyak NaN woman & mean: {lenNanWoman} & {woman}')
print(f'Banyak NaN child & mean: {lenNanChild} & {child}')
```

```python
dataTitanic['age'] = dataTitanic.groupby('who')['age'].apply(lambda x: x.fillna(x.mean()))
```

```python
lenNanMan = len(dataTitanic[(dataTitanic['who'] == 'man') & (dataTitanic['age'].isna())])
lenNanChild = len(dataTitanic[(dataTitanic['who'] == 'child') & (dataTitanic['age'].isna())])
lenNanWoman = len(dataTitanic[(dataTitanic['who'] == 'woman') & (dataTitanic['age'].isna())])
print(f'Banyak NaN man & mean: {lenNanMan} & {man}')
print(f'Banyak NaN woman & mean: {lenNanWoman} & {woman}')
print(f'Banyak NaN child & mean: {lenNanChild} & {child}')
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
classData = dataTitanic['class'].unique()
print(classData)
dataOutliers = pd.DataFrame(columns=['id', 'mean_fare', 'std_fare', 'batas_min_fare', 'batas_max_fare'], 
                           data=[[col + 1, dataTitanic[dataTitanic['class'] == classData[col]]['fare'].mean(),
                                  dataTitanic[dataTitanic['class'] == classData[col]]['fare'].std(),
                                 dataTitanic[dataTitanic['class'] == classData[col]]['fare'].mean() - (dataTitanic[dataTitanic['class'] == classData[col]]['fare'].std() *3),
                                 dataTitanic[dataTitanic['class'] == classData[col]]['fare'].mean() + (dataTitanic[dataTitanic['class'] == classData[col]]['fare'].std() *3)] 
                                 for col in range(3)])
dataOutliers
```

#### Print outlier % of each class

```python
percentageOut1 = (len(dataTitanic[((dataTitanic['fare'] < -150.986430) | (dataTitanic['fare'] > 319.295805))])/len(dataTitanic[dataTitanic['class'] == 'First']))*100
percentageOut2 = (len(dataTitanic[((dataTitanic['fare'] < -19.590013) | (dataTitanic['fare'] > 60.914379)) & (dataTitanic['class'] == 'Second')])/len(dataTitanic[dataTitanic['class'] == 'Second']))*100
percentageOut3 = (len(dataTitanic[((dataTitanic['fare'] < -21.658875) | (dataTitanic['fare'] > 49.009975))& (dataTitanic['class'] == 'Third')])/len(dataTitanic[dataTitanic['class'] == 'Third']))*100

print('Outlier % First : ' + str(percentageOut1))
print('Outlier % Second : ' + str(percentageOut2))
print('Outlier % Third : ' + str(percentageOut3))
```

#### Tambahkan column outliercalcId (FK) di dfTitanic

```python
def inputOutlierId(classes):
    if classes == 'Third':
        return 1
    elif classes == 'First':
        return 2
    return 3
```

```python
outlierId = dataTitanic['class'].apply(inputOutlierId)
dataTitanic.insert(15, 'outliercalcId', outlierId)
dataTitanic.head()
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
def checkOutlier(fare):
    if fare < (-1 * 50.986430) and fare > 319.295805 and dataTitanic['class'] == 'First':
        return 1
    if fare < (-1 * 19.590013) and fare > 60.914379 and dataTitanic['class'] == 'Second':
        return 1
    if fare < (-1 * 21.658875) and fare > 49.009975 and dataTitanic['class'] == 'Third':
        return 1
    return 0
```

```python
checkingOutlier = dataTitanic['fare'].apply(checkOutlier)
dataTitanic.insert(16, 'outlier', checkingOutlier)
dataTitanic.head()
```
